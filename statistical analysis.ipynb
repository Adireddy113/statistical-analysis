```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load dataset
df = pd.read_csv(r'C:\Users\Administrator\Downloads\student_scores.csv') 

# Basic Statistics
print("ðŸ“ˆ Statistical Summary:")
print(df.describe())

# Correlation
correlation = df.corr().iloc[0, 1]
print(f"\nCorrelation between Hours and Scores: {correlation:.2f}")

# Scatter Plot
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Hours', y='Scores', data=df, color='blue')
plt.title("Study Hours vs Student Scores")
plt.xlabel("Study Hours")
plt.ylabel("Scores")
plt.show()

# Linear Regression Model
X = df[['Hours']]
y = df['Scores']
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Regression Line
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Actual Scores')
plt.plot(X, y_pred, color='red', label='Predicted Scores')
plt.title("Regression Line - Study Hours vs Scores")
plt.xlabel("Study Hours")
plt.ylabel("Scores")
plt.legend()
plt.show()

# Evaluation Metrics
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print(f"\nðŸ“Š Mean Squared Error: {mse:.2f}")
print(f"âœ… RÂ² Score: {r2:.2f}")
```
